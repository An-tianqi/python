# Visualisation de relations continues

```python
sales = pd.read_csv('/Users/an/Downloads/sales_predictions.csv');sales
```

```python
df=pd.DataFrame(data=sales,columns=['item_price','date']);df
```

```python
# Q2
sb.relplot(x="date",y="item_price",data=sales)
```

```python
# Q3
tmp=sales.sample(n=50);
sb.relplot(x="date",y="item_price",data=tmp);
```

```python
# Q4
g = sb.relplot(x="date",y="item_price",data=tmp);
g.figure.set_size_inches(20,6);
```

```python
# Q5
tmp['date'] = pd.to_datetime(tmp['date'],infer_datetime_format=True);
a=sb.relplot(x="date",y="item_price",data=tmp);
a.figure.set_size_inches(20,6);
```

```python
b=sb.lineplot(x="date",y="item_price",data=tmp);
b.figure.set_size_inches(10,6);
```
